### Communication data available from wave 4 onwards 

#### 1. Cleaning and initial data preprocessing

```python
# taking nominations from all waves
# we need to infer the Wave variable (not included in the input files)
path = r'../data/movez_social_buzz/Buzz_per_wave' # use your path
all_files = glob.glob(os.path.join(path , "*.csv"))
appended_data = []

for f in all_files:
    print(f)
    df_com = pd.read_csv(f,encoding= 'unicode_escape',sep=';', header=0)
    nums = [int(s) for s in f if s.isdigit()]
    df_com['Wave'] = nums[0]
    appended_data.append(df_com)
    
df_com = pd.concat(appended_data)
```

```python
df_com.head()
```

```python
df_com = df_com[['Participant', 'School', 'Class', 'Child', 'Gender', 'Age', 'PI','Date', 'Time', 'UMID', 'Format_text', 'Format_Media', 'Type','RecipientChild', 'ClassReceive', 'Media', 'ReactionInto', 'Flagged','Number_likes','Wave']]
df_com.head()
```

```python
df_com.dtypes
```

##### There is some mesiness in the Number_likes labeling

```python
df_com.Number_likes.unique()
```

```python
# keep only the numeric values of this variable - the others are mistakes (and very rare)
df_com['Number_likes'] = pd.to_numeric(df_com['Number_likes'], errors='coerce')
```

```python
df_com.groupby(['Number_likes']).count()
```

```python
df_com.RecipientChild = df_com.RecipientChild.astype('Int64')
```

```python
df_com.RecipientChild = df_com.RecipientChild.astype('Int64')
```

```python
df_com.Child = df_com.Child.astype('Int64')
```

##### Removing the mymovez messages - messages that are sent from the system

```python
df_com = df_com[~df_com.Type.isin(['mymovez'])]
```

##### We will store the final outcomes in this dataframe (df_final). The NumWaves is used for averaging the outcomes.

```python
df_final = df_com.groupby('Child')['Wave'].nunique().reset_index(name='NumWaves')
df_final.head()
```

```python
df_pp = pd.read_csv('../data/movez_metadata/Participant_Info.csv', sep=';', header=0)
df_pp.shape
```

```python
df_pp.head()
```

```python
df_pp = df_pp[(df_pp['W4']==1) | (df_pp['W5']==1) | (df_pp['W6']==1) | (df_pp['W7']==1)]
df_pp =  df_pp.drop(['W1','W2','W3'], axis=1) 
df_pp['NumWavesPP'] = df_pp[['W4','W5','W6','W7']].sum(axis=1)
df_pp
```

```python
df_final['NumWavesPP'] = df_final['Child'].map(df_pp.set_index('Child')['NumWavesPP'])
df_final.NumWavesPP = df_final.NumWavesPP.astype('Int64')
df_final
```

#### 2. Deriving features from group conversations 

```python
df_com_group = df_com[df_com.Type.isin(['group'])]
df_com_group.head()
```

##### Run the next line if you want to check number of group messages exchanged on class level - which were the texting classes? Maybe check: are these the classes where influencers are detected?

```python
# df_com_group.groupby('Class')['UMID'].count().reset_index().plot.bar(x='Class', y='UMID', rot=0,figsize=(14,14))
```

##### likes are available only for group messages, and those who participated in group messages

```python
df_final['num_likes'] = df_final['Child'].map(df_com_group.groupby('Child')['Number_likes'].sum())
df_final['g_num_text'] = df_final['Child'].map(df_com_group.groupby('Child')['Format_text'].sum())
df_final['g_num_media'] = df_final['Child'].map(df_com_group.groupby('Child')['Format_Media'].count())
```

```python
df_final.head()
```

#### 3. Deriving features from one-on-one communication

```python
df_com_ind = df_com[~df_com.Type.isin(['group'])]
df_com_ind
```

```python
df_final['i_num_text'] = df_final['Child'].map(df_com_ind.groupby('Child')['Format_text'].sum())
df_final['i_num_media'] = df_final['Child'].map(df_com_ind.groupby('Child')['Format_Media'].count())
# how many individual chats were they part of - how many peers have they talked to individually?
df_final['i_num_recepients'] = df_final['Child'].map(df_com_ind.groupby('Child')['RecipientChild'].nunique())
```

```python
df_final
```

##### average the variables by dividing with the number of waves

```python
df_final['num_likes'] = df_final['num_likes']/df_final['NumWavesPP']
df_final['g_num_text'] = df_final['g_num_text']/df_final['NumWavesPP']
df_final['g_num_media'] = df_final['g_num_media']/df_final['NumWavesPP']
df_final['i_num_text'] = df_final['i_num_text']/df_final['NumWavesPP']
df_final['i_num_media'] = df_final['i_num_media']/df_final['NumWavesPP']
df_final['i_num_recepients'] = df_final['i_num_recepients']/df_final['NumWavesPP']
df_final
```

```python
df_liwc[['Child','UMID','Text','Recipient','Type','Likes','WC','affect','drives','social','polarity','health','cogproc','subjectivity']]
```

```python
df_final['WC'] = df_final['Child'].map(df_final2.set_index('Child')['WC'])
df_final['affect'] = df_final['Child'].map(df_final2.set_index('Child')['affect'])
df_final['drives'] = df_final['Child'].map(df_final2.set_index('Child')['drives'])
df_final['social'] = df_final['Child'].map(df_final2.set_index('Child')['social'])
df_final['polarity'] = df_final['Child'].map(df_final2.set_index('Child')['polarity'])
df_final['health'] = df_final['Child'].map(df_final2.set_index('Child')['health'])
df_final['cogproc'] = df_final['Child'].map(df_final2.set_index('Child')['cogproc'])
df_final['subjectivity'] = df_final['Child'].map(df_final2.set_index('Child')['subjectivity'])
df_final
```
